```python
import os
import re
from nltk.tokenize import word_tokenize
```

```python
with open("AllCombined.txt","r",encoding="utf-8") as f:
  lines = f.readlines()
```

```python
clean_text  = [line.strip() for line in lines if line.strip() != ""]
```

```python
sentences = []
```

```python
for text in clean_text:
  text = text.lower()
  text = re.sub(r"[^a-z\s]", "", text)
  tokens = word_tokenize(text)
  sentences.append(tokens)
```

```python
from gensim.model import Word2Vec
```

```python
pip install gensim
```

```python
model = Word2Vec(sentences = sentences,vector_size = 100,window= 5,min_count = 5,sg=1)
```

```python
model.save('optimizer.model')
```

```python
model.wv.most_similar("science")
```
